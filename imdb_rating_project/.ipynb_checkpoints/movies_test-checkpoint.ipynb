```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("imdb_clean.csv")
```

```python
#Top rated movie for each year from 2001-2023
current_year = pd.Timestamp.now().year
start_year = current_year - 22
filtered_data = data[data['release_year'].between(start_year, current_year)]
top_movies = filtered_data.groupby('release_year')['rating'].idxmax()
top_movies_data = filtered_data.loc[top_movies]
top_movies_table = top_movies_data[['release_year', 'title', 'rating']]
top_movies_table.columns = ['Year', 'Movie', 'Rating']
top_movies_table.reset_index(drop=True, inplace=True)
top_movies_table
```

```python
# Top rating for each year
current_year = pd.Timestamp.now().year
start_year = current_year - 22
filtered_data = data[data['release_year'].between(start_year, current_year)]

top_movies = filtered_data.groupby('release_year')['rating'].idxmax()
top_movies_data = filtered_data.loc[top_movies]

plt.figure(figsize=(10, 6))
bars = plt.bar(top_movies_data['release_year'].astype(str), top_movies_data['rating'])
plt.xlabel('Year')
plt.ylabel('Rating')
plt.title('Top-rated Movie by Year (Last 23 years)')
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Count the number of movies released per year that are in the top 1000
movie_counts = data['release_year'].value_counts().sort_index()

plt.figure(figsize=(25, 6))
bars = plt.bar(movie_counts.index.astype(str), movie_counts.values)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Released per Year that are in the top 1000')
plt.xticks(rotation=45)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, str(yval), ha='center', va='bottom')


plt.tight_layout()
plt.show()
```

```python
# Top rating for each year
current_year = pd.Timestamp.now().year
start_year = current_year - 22
filtered_data = data[data['release_year'].between(start_year, current_year)]

top_movies = filtered_data.groupby('release_year')['gross(M)'].idxmax()
top_movies_data = filtered_data.loc[top_movies]

plt.figure(figsize=(10, 6))
bars = plt.bar(top_movies_data['release_year'].astype(str), top_movies_data['gross(M)'])
plt.xlabel('Year')
plt.ylabel('Total Gross in Millions')
plt.title('Top-rated Movie by Year (Last 23 years)')
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
